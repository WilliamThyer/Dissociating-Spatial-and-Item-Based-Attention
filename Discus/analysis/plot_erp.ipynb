```python
import sys
sys.path.append('../..')

import eeg_decoder
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
%load_ext autoreload
%autoreload 2
```

```python
experiment_name = 'd01'
data_dir = f'./data/{experiment_name}'
exp = eeg_decoder.Experiment(experiment_name,data_dir,dev=False,trim_timepoints=[-200,1000])
erp = eeg_decoder.ERP(exp,subtitle=experiment_name)
xdata_all, ydata_all = erp.load_all_eeg()
```

```python
fig,axes = plt.subplots(1,3,figsize=(19,6))
for isubset,(subtitle,subset) in enumerate(zip(['_ssPO','_ssC','_ssF'],[('P','O'),('C'),('F')])):
    erp.plot_ss(
        xdata_all, ydata_all,
        ax = axes[isubset], 
        electrode_subset=subset,
        condition_labels = ['1 (very dissimilar)','1 (dissimilar)','1 (similar)','2'])
    
plt.show()
```
